# Lol api calls

Define api  ->
Criar env com a key

```python
import requests
import os   
from dotenv import load_dotenv
api_key = os.getenv("RIOT_API_KEY")
```

Get name and region and get puuid

```python
summoner_name = "GoncasCC"
summoner_tag = "EUW"
region = "europe"

search_summoner_url = "https://" + region + ".api.riotgames.com/riot/account/v1/accounts/by-riot-id/" + summoner_name + "/" + summoner_tag + "?api_key="+ api_key    

resp = requests.get(search_summoner_url)
player_info = resp.json()
puuid = player_info['puuid']
print("puuid: " + puuid)
```

Get latest matches 

```python
search_matches_url = "https://" + region + ".api.riotgames.com/lol/match/v5/matches/by-puuid/" + puuid + "/ids?start=0&count=20&api_key=" + api_key
resp  = requests.get(search_matches_url)
player_matches = resp.json()
last_match_id = player_matches[0]
print("last match id: " + last_match_id)
```

Get information about laast match

```python
search_last_match_info_url = "https://" + region + ".api.riotgames.com/lol/match/v5/matches/" + last_match_id + "?api_key=" + api_key
resp = requests.get(search_last_match_info_url)
resp.json()
```
